```python
### The labels of the Fashion MNIST dataset. We just plot a few images to get a very rough idea of what some 
### data samples look like.

labels_map = {
    0: "T-Shirt",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle Boot",
}
figure = plt.figure(figsize=(8, 8))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(len(train_data), size=(1,)).item()
    img, label = train_data[sample_idx]
    figure.add_subplot(rows, cols, i)
    plt.title(labels_map[label])
    plt.axis("off")
    plt.imshow(img.squeeze(), cmap="gray")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# First plot
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)

plt.figure()  # Create a new figure
plt.plot(x1, y1, label='sin(x)')
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()

# Second plot
x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)

plt.figure()  # Create another new figure
plt.plot(x2, y2, label='cos(x)', color='red')
plt.title('Cosine Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def append_bias(X):
    """
    TODO
    Appends bias to the input
    args:
        X (N X d 2D Array)
    returns:
        X_bias (N X (d+1)) 2D Array
    """
    bias_column = np.ones((X.shape[0], 1))
    X_bias = np.hstack((X, bias_column))
    
    return X_bias

append_bias(np.array([[2, 2],[2, 2]]))
```
